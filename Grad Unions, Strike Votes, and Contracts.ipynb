```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('grad_union_data.csv')
df = df[df.got_contract == 1]
df
```

```python
plt.rcParams.update({'font.size': 16})
```

```python
plt.figure(figsize=(10,5))
xes = np.arange(2)

got_protections =  [(df[df.sav==1].got_strong_protections==1).sum(),
                    (df[df.sav==0].got_strong_protections==1).sum()]

no_protections = [(df[df.sav==1].got_strong_protections==0).sum(),
                  (df[df.sav==0].got_strong_protections==0).sum()]

plt.bar(xes, no_protections, label='No 3rd-party arbitration')
plt.bar(xes, got_protections, bottom=no_protections, label='3rd-party arbitration! Woot!')
plt.legend(loc='upper left')
plt.xticks(xes, ['SAV', 'No SAV'])
plt.ylabel('Number of grad student unions')
plt.title("Holding SAVs is correlated with getting better contracts")
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.title("Not really enough data to make a histogram, but w/e")
plt.hist(df[df.sav==1].strike_days, bins=np.arange(23)-0.5)
plt.xticks(np.arange(22))
plt.xlabel("Number of days on strike")
plt.ylabel("Number of unions")
plt.show()
```
